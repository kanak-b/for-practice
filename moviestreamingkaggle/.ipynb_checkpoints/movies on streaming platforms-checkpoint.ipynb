```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.drop(['Type', 'Unnamed: 0', 'ID'], axis = 1, inplace = True)
```

```python
df.isna().sum()
```

```python
df.drop(['Age','Rotten Tomatoes'], axis = 1, inplace =True)
df.dropna(how='any', axis = 0, inplace =True)
df.isna().sum().sum()
```

```python
df.shape
```

```python
#!pip install sidetable
import sidetable
df.stb.freq(['Language'], thresh=.8)
```

```python
df['IMDb'].describe()
```

```python
plt.figure(figsize=(10,6))
plt.title("Distribution of IMDb Ratings", fontsize=15)
sns.distplot(df['IMDb'])
```

```python
df[['IMDb','Directors']].groupby('Directors').agg(['mean','count']).sort_values(by=('IMDb','mean'), ascending=False)[:10]
```

```python
directors = df[['IMDb','Directors']].groupby('Directors').agg(['mean','count'])
directors[directors[('IMDb', 'count')] > 5].sort_values(by=('IMDb','mean'), ascending=False)[:10]
```

```python
df[['IMDb','Netflix']].groupby(['Netflix']).agg(['mean','count'])
```

```python
streams = ['Netflix', 'Hulu', 'Prime Video', 'Disney+']
dict_a = {}
for stream in streams:
    a = df[['IMDb', stream]].groupby([stream]).agg(['mean',
        'count']).iloc[1,:].values
    dict_a[stream] = a
    df_stream = pd.DataFrame(dict_a, 
                index=['Avg_IMDb','Number of Moviews'])
df_stream
```
